To perform a join, SQL uses **shared columns** between tables. These columns are usually defined using **primary keys and foreign keys**.

To analyze or generate reports, we often need information from **multiple tables at once**. This is done using **JOIN statements**.


# Joining Product and Order Data

Suppose we want to generate a report showing **product information together with order details**.

Since the data is stored in two tables (`orderdetails` and `products`), we must **join them together**.

```sql
SELECT *
FROM orderdetails
JOIN products
ON orderdetails.productCode = products.productCode
LIMIT 10;
```

### What this query does

- Combines the `orderdetails` and `products` tables
- Matches rows where `productCode` is the same
- Returns the **first 10 results**


````python
%%sql
SELECT *
FROM orderdetails
JOIN products
ON orderdetails.productCode = products.productCode
LIMIT 3;
````

# The USING Clause

If the **column name used for joining is identical in both tables**, SQL provides a simpler syntax using the **USING clause**.

Instead of writing:

```sql
ON tableA.column = tableB.column
```

We can write:

```sql
USING(column)
```

# Example Using the USING Clause

```sql
SELECT *
FROM orderdetails
JOIN products
USING(productCode)
LIMIT 10;
```
This query performs the same join as before but with **shorter and cleaner syntax**.

````python
%%sql
SELECT *
FROM orderdetails
JOIN products
USING(productCode)
LIMIT 2;
````

# SQL JOIN Relationships

## One-to-One Relationships

In a **one-to-one relationship**, each record in one table is associated with **exactly one record in another table**.

When performing a JOIN in this case, the **number of rows does not increase**.  
Instead, the query simply **adds more columns of information**.


## Cities for Sales Rep Employees

Now join the **employees** table with the **offices** table to include the city where the employee works.

```sql
SELECT firstName, lastName, email, city
FROM employees
JOIN offices
USING(officeCode)
WHERE jobTitle = 'Sales Rep';
```

Result:

- The **number of rows remains the same**
- The query simply adds the **city column**


````python
%%sql
SELECT firstName, lastName, email, city
FROM employees
JOIN offices
USING(officeCode)
WHERE jobTitle = 'Sales Rep'
LIMIT 3;
````

# One-to-Many Relationships

A **one-to-many relationship** occurs when:

- One record in a table is associated with **multiple records in another table**

In this case, a JOIN will **increase the number of rows** in the result.


## Joining Product Lines with Products

Now join the **productlines** table with the **products** table.

```sql
SELECT productLine, textDescription, productVendor, productDescription
FROM productlines
JOIN products
USING(productLine);
```

Result:

- Each **product line appears multiple times**
- Once for **each product within that product line**

This increases the **number of rows in the result**.



````python
%%sql
SELECT productLine, textDescription, productVendor, productDescription
FROM productlines
JOIN products
USING(productLine)
LIMIT 2;
````

# Many-to-Many Relationships

A **many-to-many relationship** occurs when:

- Multiple records in one table relate to **multiple records in another table**

This can dramatically **increase the number of rows** produced by a JOIN.

Example:

- 2 offices in a state
- 9 customers in the same state

Joining by state would produce:

```
2 × 9 = 18 rows
```

## Joining Offices and Customers

Joining these tables on **state** would produce a many-to-many result.

```sql
SELECT
    offices.officeCode,
    offices.city,
    offices.state,
    customers.customerName,
    customers.city AS customerCity
FROM offices
JOIN customers
ON offices.state = customers.state
ORDER BY offices.state;
```

### What Happens in This Query

- Every **office in a state** is matched with **every customer in that same state**
- If there are multiple offices and customers in the same state, the rows multiply

````python
%%sql
SELECT
    offices.officeCode,
    offices.city,
    offices.state,
    customers.customerName,
    customers.city AS customerCity
FROM offices
JOIN customers
ON offices.state = customers.state
ORDER BY offices.state LIMIT 3
````
